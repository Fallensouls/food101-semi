```python

import numpy as np
import cv2
import os
 
# img_h, img_w = 32, 32
img_h, img_w = 64, 64   #根据自己数据集适当调整，影响不大
means=np.zeros((1,3))
stdevs=np.zeros((1,3))
 
folder_path = './food-101/images'
folder_path_list = os.listdir(folder_path)
 
len_ = len(folder_path_list)
i = 0
for item in folder_path_list:
    img_list = []
    image_folder_path = os.path.join(folder_path,item)
    for it in os.listdir(image_folder_path):
        img = cv2.imread(os.path.join(image_folder_path,it))
        # print(os.path.join(folder_path,item,it))
        img = cv2.resize(img,(img_w,img_h))
        img = img[:, :, :, np.newaxis]
        img_list.append(img)
        i += 1
        if i%1000 == 0:
            print(i,'/',len_*1000)    
 
    imgs = np.concatenate(img_list, axis=3)
    imgs = imgs.astype(np.float32) / 255.
    
    for i in range(3):
        pixels = imgs[:, :, i, :].ravel()  # 拉成一行
        means[0,i]+=(np.mean(pixels))
        stdevs[0,i]+=(np.std(pixels))
    print(means[:,[2,1,0]],stdevs[:,[2,1,0]])
    # BGR --> RGB ， CV读取的需要转换，PIL读取的不用转换

print(means[:,[2,1,0]],stdevs[:,[2,1,0]])
 
print("normMean = {}".format(means))
```
